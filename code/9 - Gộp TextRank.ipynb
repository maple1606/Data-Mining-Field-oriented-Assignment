```python
import pandas as pd

combined_keywords = pd.concat([df_ner[['idx', 'keywords']], combined_df[['idx', 'keywords']]])
combined_keywords = combined_keywords.drop_duplicates(subset='idx', keep='last')
combined_keywords_sorted = combined_keywords.sort_values(by='idx', ascending=True)

print(combined_keywords.head())
```

```python
import pandas as pd

combined_keywords_grouped = combined_keywords.groupby('idx', as_index=False).agg({
    'keywords': lambda x: ','.join(filter(None, map(str, x)))
})
combined_keywords_grouped['keywords'] = combined_keywords_grouped['keywords'].fillna('').str.replace('_', ' ', regex=False)
combined_keywords_grouped['keywords'] = combined_keywords_grouped['keywords'].fillna('').str.lower()
print(combined_keywords_grouped)
```

```python
def remove_duplicate_keywords(keywords):
    if pd.isna(keywords):
        return keywords

    keywords_list = keywords.split(',')
    unique_keywords = list(dict.fromkeys(keywords_list))
    return ','.join(unique_keywords)

combined_keywords_grouped['keywords'] = combined_keywords_grouped['keywords'].apply(remove_duplicate_keywords)

print(combined_keywords_grouped)
```

```python
merged_df = pd.merge(df, combined_keywords_grouped, on='idx', how='left')
merged_df = merged_df.T.drop_duplicates().T

print(merged_df)
```

```python
import numpy as np

merged_df['keywords'] = merged_df['keywords_x'].fillna('') + ', ' + merged_df['keywords_y'].fillna('')
merged_df['keywords'] = merged_df['keywords'].str.strip(', ')
merged_df['keywords'] = merged_df['keywords'].replace('', np.nan)

merged_df.drop(columns=['keywords_x', 'keywords_y'], inplace=True)

print(merged_df)
```
