# Diễn giải

```python
import pandas as pd

output_file_name = "/content/drive/MyDrive/Data Mining Field-oriented/stats/diễn giải chủ đề.csv"

labels_map = {
    "Chủ đề 0": "Thể thao Giải trí",
    "Chủ đề 1": "Kinh doanh",
    "Chủ đề 2": "Kế hoạch đầu tư",
    "Chủ đề 3": "Khí tượng Sức khỏe",
    "Chủ đề 4": "Pháp luật",
    "Chủ đề 5": "Đời sống Giải trí",
    "Chủ đề 6": "Thế giới",
    "Chủ đề 7": "Thời sự",
    "Chủ đề 8": "Ngoại giao",
    "Chủ đề 9": "Giáo dục",
    "Chủ đề 10": "Sức khỏe Công nghệ",
    "Chủ đề 11": "Xã hội",
}

df['Chủ đề 1'] = df['Chủ đề 1'].str.extract(r"(Chủ đề \d+)")[0]
print(df['Chủ đề 1'])
df['Chủ đề 1'] = df['Chủ đề 1'].map(labels_map)
df.rename(columns={"Chủ đề 1": "Chủ đề chính"}, inplace=True)

df.to_csv(output_file_name, index=False, encoding="utf-8-sig")
```

```python
output_folder = "/content/drive/MyDrive/Data Mining Field-oriented/data/consolidated/topic"
df_grouped = df.groupby("Chủ đề chính")

print(df_grouped)

file_paths = {}
for group_name, group_data in df_grouped:
    file_name = f"{output_folder}/{group_name.replace(' ', '_')}.csv"
    group_data.to_csv(file_name, index=False, encoding="utf-8-sig")
    file_paths[group_name] = file_name
```
